```python
import os 
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords

pd.set_option('display.max_colwidth', 100)
stopwords =  nltk.corpus.stopwords.words('english')
df = pd.read_csv(r'C:\Users\ghect\Documents\Bootcamp\Proyectos\Data_Cleaning\Datos_Kaggle\Noticias_falsas_crudo.csv', encoding="ISO-8859-1")
df = df.set_index('id', drop=True)
df.head()
```

# Eliminar columnas

```python
def eliminar_nulos(df):
    null_cols = df.isnull().sum()
    eliminar_cols = list(null_cols[null_cols >= 4600].index)
    df_nulos = df.drop(eliminar_cols, axis=1)
    df_nulos = df_nulos.drop('Unnamed: 5', axis=1)
    df_eliminar_nulos = df_nulos.dropna()
    return df_eliminar_nulos

df_sin_nulos = eliminar_nulos(df)
df_sin_nulos.head()
```

# Limpiar texto

```python
def limpiar_texto(txt):
    txt = "".join([caracter for caracter in txt if caracter not in string.punctuation])
    tokens = re.split('\W+', txt)
    txt = [palabra for palabra in tokens if palabra not in stopwords]
    return txt

def texto_limpio_aplicado(df_sin_nulos):
    df_sin_nulos['autor_limpio'] = df_sin_nulos['author'].apply(lambda x: limpiar_texto(x.lower()))
    df_sin_nulos['titulo_limpio'] = df_sin_nulos['title'].apply(lambda x: limpiar_texto(x.lower()))
    df_sin_nulos['texto_limpio'] = df_sin_nulos['text'].apply(lambda x: limpiar_texto(x.lower()))
    df_texto_limpio = df_sin_nulos.loc[:, ['titulo_limpio', 'autor_limpio', 'texto_limpio', 'label']]
    return df_texto_limpio

df_limpio = texto_limpio_aplicado(df_sin_nulos)
df_limpio.head()
```

# Agregar una columna de variables

```python
def clas(var: int) -> int :
  """
  Recibe un valor y lo evalua como numero.

  Regresa un entero
    INT
  """

  if var == 'true':
    return 0
  else:
    return 1

df_nuevo = df_limpio.rename(columns={'titulo_limpio': 'titulo', 'autor_limpio': 'autor', 'texto_limpio': 'texto'})
df_nuevo['clasificacion'] = df_nuevo['label'].apply(clas)
df_nuevo.head()
```
